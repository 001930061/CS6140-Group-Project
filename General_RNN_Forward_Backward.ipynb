### Part 3: Model Definition and Supporting Functions

#### Purpose:  
The purpose of this section is to define the deep learning model used for time series forecasting, specifically predicting stock prices or other sequential data. The model is based on unidirectional RNN (Recurrent Neural Network) layers and a Dense layer for the final prediction. It also includes a custom accuracy calculation function and a smoothing function for better prediction results.

#### Significance:  
- **Model Architecture**: The model uses a unidirectional RNN network to process data in a single direction, capturing temporal dependencies in sequential data. This design is effective for time series forecasting, where past information is used to predict future values.
- **Accuracy Calculation**: The custom accuracy function calculates the relative accuracy between the predicted and real values, providing insight into the model's performance. This is important for monitoring the model's predictions and adjusting for better results.
- **Smoothing Function**: The anchor function applies exponential smoothing to the predicted values, improving the model's ability to generalize and reduce noise in its predictions. This is useful for making more stable and reliable forecasts.


```python
import tensorflow as tf
import numpy as np

# Define a general Bidirectional RNN model class
class Model(tf.keras.Model):
    def __init__(self, learning_rate, num_layers, size_layer, output_size, forget_bias=0.1):
        """
        Initializes the model with Bidirectional RNN layers and a dense output layer.

        :param learning_rate: Learning rate for the optimizer.
        :param num_layers: Number of RNN layers.
        :param size_layer: Number of neurons in each RNN layer.
        :param output_size: Dimension of the output layer.
        :param forget_bias: Dropout rate to prevent overfitting.
        """
        super(Model, self).__init__()  # Call the parent constructor
        self.size_layer = size_layer  # Number of neurons in each RNN layer
        self.num_layers = num_layers  # Number of RNN layers
        
        # Define Bidirectional RNN layers with dropout for each layer
        self.rnn_cells = [
            tf.keras.layers.Bidirectional(tf.keras.layers.SimpleRNN(size_layer, return_sequences=True, dropout=forget_bias))
            for _ in range(num_layers)
        ]
        
        # Fully connected layer for the final prediction
        self.dense = tf.keras.layers.Dense(output_size)  
        
        # Adam optimizer
        self.optimizer = tf.keras.optimizers.Adam(learning_rate)

    def call(self, X_forward, X_backward):
        """
        Forward and backward pass through the Bidirectional RNN layers.

        :param X_forward: Input tensor for the forward pass.
        :param X_backward: Input tensor for the backward pass.
        :return: Predicted logits.
        """
        # Forward pass through Bidirectional RNN layers
        forward_output = X_forward
        for cell in self.rnn_cells:
            forward_output = cell(forward_output)
        
        # Backward pass through Bidirectional RNN layers
        backward_output = X_backward
        for cell in self.rnn_cells:
            backward_output = cell(backward_output)

        # Compute the difference between backward and forward outputs
        output = backward_output - forward_output
        logits = self.dense(output[:, -1, :])  # Use the last time step's output for the final prediction
        return logits

# Accuracy calculation function
def calculate_accuracy(real, predict):
    """
    Calculates the accuracy of predictions.

    :param real: Array of true values.
    :param predict: Array of predicted values.
    :return: Accuracy percentage.
    """
    real = np.array(real) + 1  # Shift real values to positive range
    predict = np.array(predict) + 1  # Shift predicted values to positive range
    # Calculate accuracy using error formula
    percentage = 1 - np.sqrt(np.mean(np.square((real - predict) / real)))  
    return percentage * 100  # Return accuracy as a percentage

# Smoothing function
def anchor(signal, weight):
    """
    Applies exponential smoothing to a signal.

    :param signal: Input signal to be smoothed.
    :param weight: Smoothing weight (0 < weight < 1).
    :return: Smoothed signal as a list.
    """
    buffer = []  # Stores the smoothed signal
    last = signal[0]  # Initialize with the first value
    for i in signal:
        # Compute the smoothed value using the exponential smoothing formula
        smoothed_val = last * weight + (1 - weight) * i
        buffer.append(smoothed_val)  # Append the smoothed value to the buffer
        last = smoothed_val  # Update the last value for the next iteration
    return buffer  # Return the smoothed signal
```

### Part 5: Model Training and Prediction Functions

#### Purpose:
This part defines four essential functions—`create_model`, `train_model`, `predict`, and `forecast`—that facilitate the creation, training, and prediction process for a General RNN-based neural network model used in time-series forecasting. Each function plays a specific role: `create_model` initializes the model with the desired hyperparameters; `train_model` trains the model over multiple epochs while tracking performance metrics; `predict` generates both historical and future predictions after training; and `forecast` ties everything together by orchestrating the complete pipeline, from model creation to forecasting.

#### Significance:
1. **`create_model`**:
   - This function is responsible for constructing the General RNN model, setting it up with key parameters such as the learning rate, number of layers, and dropout rate. The proper initialization of the model is critical for ensuring that the network is structured to handle the time-series data effectively.

2. **`train_model`**:
   - The `train_model` function is responsible for training the model through multiple epochs. During each epoch, the function processes batches of data, computes the loss (Mean Squared Error), and updates the model’s weights using backpropagation. The function also tracks key performance metrics like loss and accuracy, helping to ensure that the model is learning in an optimal way.

3. **`predict`**:
   - The `predict` function generates predictions using the trained model. It makes predictions on both the training data and future unseen data. This function is designed to handle edge cases, such as incomplete batches, and processes the predictions by inverse scaling and smoothing, ensuring the final output is accurate and useful for forecasting future values.

4. **`forecast`**:
   - The `forecast` function ties the entire model training and prediction process together. It creates the model, trains it on the provided training data, and generates future predictions. This function serves as the entry point for the entire pipeline and ensures that all components are executed in sequence, yielding the final forecasted values.

Together, these functions form a complete system for training a General RNN-based model, generating predictions, and performing time-series forecasting, which is essential for tasks like predicting stock prices, weather patterns, or other temporal data.


```python
def create_model():
    """
    Initialize and clear the TensorFlow session to create a fresh model.

    Returns:
        Model: The initialized custom Bidirectional GRU-based model.
    """
    K.clear_session()  # Clear the current TensorFlow session to avoid conflicts.

    # Initialize the custom model with predefined parameters.
    modelnn = Model(learning_rate, num_layers, size_layer, df_log.shape[1], dropout_rate)
    
    return modelnn

def train_model(modelnn, df_train, epoch, timestamp):
    """
    Train the custom model using the training dataset.

    Args:
        modelnn (Model): The Bidirectional GRU-based model to train.
        df_train (DataFrame): The training dataset containing time-series data.
        epoch (int): Number of training epochs.
        timestamp (int): Length of the input sequence for each batch.

    Returns:
        Model: The trained model.
    """
    pbar = tqdm(range(epoch), desc='Training Loop')  # Display a progress bar for training.
    
    for i in pbar:
        # Initialize the forward and backward state arrays for the Bidirectional GRU model.
        init_value_forward = np.zeros((1, num_layers * 2 * size_layer))
        init_value_backward = np.zeros((1, num_layers * 2 * size_layer))
        
        total_loss, total_acc = [], []  # Track loss and accuracy across batches.

        for k in range(0, df_train.shape[0] - 1, timestamp):
            index = min(k + timestamp, df_train.shape[0] - 1)
            
            # Prepare the input data for forward and backward passes.
            batch_x_forward = np.expand_dims(df_train.iloc[k: index, :].values, axis=0)
            batch_x_backward = np.expand_dims(np.flip(df_train.iloc[k: index, :].values, axis=0), axis=0)
            batch_y = df_train.iloc[k + 1: index + 1, :].values

            # Compute gradients and update model weights.
            with tf.GradientTape() as tape:
                logits = modelnn(batch_x_forward, batch_x_backward)
                mse_loss = tf.keras.losses.MeanSquaredError()
                loss = mse_loss(batch_y, logits)

            gradients = tape.gradient(loss, modelnn.trainable_variables)
            modelnn.optimizer.apply_gradients(zip(gradients, modelnn.trainable_variables))

            total_loss.append(loss.numpy())
            total_acc.append(calculate_accuracy(batch_y[:, 0], logits[:, 0]))

        # Update the progress bar with the latest loss and accuracy.
        pbar.set_postfix(cost=np.mean(total_loss), acc=np.mean(total_acc))

    return modelnn

def predict(modelnn, df_train, test_size, timestamp):
    """
    Generate predictions using the trained model.

    Args:
        modelnn (Model): The trained Bidirectional GRU-based model.
        df_train (DataFrame): The training dataset for reference during prediction.
        test_size (int): Number of future time steps to predict.
        timestamp (int): Length of the input sequence for each batch.

    Returns:
        np.array: The predicted future values.
    """
    # Initialize the output array for storing predictions.
    future_day = test_size
    output_predict = np.zeros((df_train.shape[0] + future_day, df_train.shape[1]))
    output_predict[0] = df_train.iloc[0]

    upper_b = (df_train.shape[0] // timestamp) * timestamp

    # Predict on the training dataset.
    for k in range(0, (df_train.shape[0] // timestamp) * timestamp, timestamp):
        batch_x_forward = np.expand_dims(df_train.iloc[k: k + timestamp, :], axis=0)
        batch_x_backward = np.expand_dims(np.flip(df_train.iloc[k: k + timestamp, :].values, axis=0), axis=0)
        out_logits = modelnn(batch_x_forward, batch_x_backward)
        output_predict[k + 1: k + timestamp + 1, :] = out_logits

    if upper_b != df_train.shape[0]:
        batch_x_forward = np.expand_dims(df_train.iloc[upper_b:, :], axis=0)
        batch_x_backward = np.expand_dims(np.flip(df_train.iloc[upper_b:, :].values, axis=0), axis=0)
        out_logits = modelnn(batch_x_forward, batch_x_backward)
        output_predict[upper_b + 1: df_train.shape[0] + 1] = out_logits
        future_day -= 1

    # Predict future values beyond the training dataset.
    date_ori = pd.to_datetime(df.iloc[:, 0]).tolist()
    for i in range(future_day):
        o = output_predict[-future_day - timestamp + i: -future_day + i]
        o_f = np.flip(o, axis=0)
        out_logits = modelnn(np.expand_dims(o, axis=0), np.expand_dims(o_f, axis=0))
        output_predict[-future_day + i] = out_logits[-1]
        date_ori.append(date_ori[-1] + timedelta(days=1))

    # Post-process predictions (inverse scaling and smoothing).
    output_predict = minmax.inverse_transform(output_predict)
    deep_future = anchor(output_predict[:, 0], 0.3)

    return deep_future[-test_size:]

def forecast():
    """
    Train the model and forecast future values.

    Returns:
        np.array: The predicted future values.
    """
    modelnn = create_model()  # Create the Bidirectional GRU-based model.
    modelnn = train_model(modelnn, df_train, epoch, timestamp)  # Train the model.
    forecast_result = predict(modelnn, df_train, test_size, timestamp)  # Generate predictions.

    return forecast_result
```

### Part 6: Simulation and Performance Evaluation

#### Purpose:
The purpose of this section is to run multiple simulations of the forecasting model using a General RNN and evaluate its performance in terms of prediction accuracy and execution time. The code runs a loop where the `forecast` function is called multiple times, and for each simulation, the prediction results are stored, and the time taken for each simulation is recorded. This allows for performance tracking and comparison across different runs of the model.

#### Significance:
This part is essential for assessing the efficiency and scalability of the forecasting model using a General RNN. The key elements are:

- **Simulation Loop**: The loop runs the forecasting process multiple times (specified by `simulation_size`). Each iteration of the loop represents a separate simulation where the General RNN model makes a prediction using the `forecast` function. This enables us to assess the model's consistency and reliability by averaging results from multiple runs.

- **Performance Tracking**: The time taken for each simulation is recorded using the `time` module, providing insight into how long each forecast takes to compute. By tracking this duration, we can assess whether the General RNN model is computationally efficient or if further optimizations are needed, especially for larger datasets.

- **Results Storage**: The `results` list stores the forecasted values for each simulation, allowing for post-simulation analysis. These results can be further analyzed to evaluate the model's accuracy or compared to ground truth data if available.

- **Efficiency Evaluation**: By printing the duration of each simulation, the process provides immediate feedback on the computational time per simulation. This helps to identify potential bottlenecks and allows us to make adjustments to improve model performance or reduce execution time.



```python
# Initialize an empty list to store the results of each simulation's forecast
results = []

# Initialize an empty list to store the duration of each simulation run (in seconds)
simulation_times = []

# Start a loop to run multiple simulations (simulation_size defines how many times to run)
for i in range(simulation_size):
    start_time = time.time()  # Record the current time as the start time of the simulation
    print(f"simulation {i + 1} started")  # Print a message indicating the start of the simulation

    # Call the forecast() function to perform the forecast and store the result in forecast_result
    forecast_result = forecast()

    # Append the forecast result to the results list for later use or analysis
    results.append(forecast_result)

    end_time = time.time()  # Record the current time as the end time of the simulation
    simulation_duration = end_time - start_time  # Calculate the time it took to complete this simulation (in seconds)

    # Append the simulation duration to the simulation_times list for later analysis
    simulation_times.append(simulation_duration)

    # Print a message indicating that the simulation has completed, along with the time it took to run
    print(f"simulation {i + 1} completed in {simulation_duration:.2f} seconds")
```
